# Конспекты по «Цифровой связи»

Представлено конспективное изложение основных моментов «Цифровой связи» — дисциплины, которая может скрываться под разными названиями, такими как «Общая теория связи», «Передача дискретных сообщений», «Теория и техника передачи информации», «Формирование и обработка сигналов систем связи» и т. п.

Целью конспектов поставлено изложение материала в достаточно общей форме без глубокого погружения в детали, что даст изучающему фундамент, который позволит изучать более детальные (но и менее доступные для понимания) книги, учебные пособия и статьи.

Данное изложение является результатом переосмысления 9-летнего опыта преподавания данной дисциплины в техническом вузе.

Материал изложен в виде относительно небольшого количества абзацев, разделенных подзаголовками, что позволяет быстро отыскивать материал в случае надобности, а также сосредотачиваться на так называемых «моментах» и осмысливать их дольше, нежели они были бы абзацами достаточно большой главы.

Базовыми требованиями для освоения данных конспектов являются знание высшей математики и теории вероятностей технического вуза, а также школьной физики. Будет плюсом знание дисциплин «Радиотехнические цепи и сигналы», «Основы теории цепей» или подобных.

# Введение

## Цифровая связь как транспортный сервис.

Под цифровой связью подразумевают техническую передачу символов с помощью некоторых сигналов через определённые каналы связи. Символами называют элементы заранее определённого алфавита, при этом алфавиты не являются алфавитами в традиционном понимании, — алфавитами являются таблицы соответствия между двоичными словами (как правило фиксированной длины) и сигналами определенной формы — информационными импульсами. Именно эта формальность подчеркивает техническую сторону цифровой связи как некоторого транспортного сервиса, которому безразлично содержание передаваемой информации.

## Канал связи. Качество канала связи. Надежная связь.

Под каналом связи понимают согласованный набор технических средств, который позволяет задавая определённое входное воздействие наблюдать связанный с ним отклик — выходное воздействие. Чем лучше мы можем судить о входном воздействии по наблюдаемому отклику, тем лучше качество канала связи. Для поддержания заданного качества тратится некоторая энергия. Чем выше требуемое качество, тем больше тратится энергии на передачу/приём одного бита (бит - двоичный символ), однако, основная соль в том, что степень роста затрачиваемой энергии от качества зависит от конкретного канала связи; стремятся, чтобы рост качества не сопровождался бы драматическим ростом потребляемой энергии.

Количественно качество канала связи в цифровых системах передачи информации определяется вероятностью ошибки наблюдаемого (принимаемого) символа. Чем качественнее канал, тем меньше повторных запросов потребуется для обеспечения надёжной связи, при этом под надёжной связью понимают практически идеальную передачу информации, когда итоговой вероятностью ошибки можно пренебречь в рамках решаемой задачи.

## Линия связи.

Под линией связи понимают некоторую физическую среду передачи, которая направляет сигнал, несущий передаваемую информацию. Направление сигнала необходимо как для экономии энергии, так и для разграничения связи по каналам, чтобы не мешать параллельно существующим системам связи/абонентам. Примерами направляющих сред могут быть любые кабели (телефонный кабель типа «лапша», витая пара, коаксиальный кабель, оптическое волокно), СВЧ-линии, волноводы, водная и воздушная среды. В случае водной и воздушной сред направленность сама по себе слабая, однако, она создается антенными системами, которые можно отнести к направляющим средам. Линия связи входит в канал связи (вложена в него) и не может быть из него исключена принципиально. Канал связи - понятие растяжимое и может включать в себя (помимо линии связи) модемы, скремблеры, кодеки, шифраторы и т.п. Канал связи задаётся в зависимости от текущей решаемой задачи (текущего уровня анализа или синтеза).

## Непрерывность сигнала, дискретность цифровой информации.

Под сигналом понимают некоторую величину, которая изменяется с течением времени. В системах связи часто это напряжение или ток в электрической цепи, интенсивность света в оптическом волокне, комплексная амплитуда электромагнитного поля в СВЧ-линии. Этой изменяющейся величине, в итоге, соответствует некоторая информация.

В системах цифровой связи информация является дискретной по своей природе (т.е. счётной); сам же сигнал всегда непрерывен по своей природе безотносительно к типу рассматриваемой системы. В аналоговых системах связи информацией является непрерывный сигнал, полученный после всех преобразований в приёмнике. Неотъемлемой частью такого сигнала является шум, который можно снижать, но нельзя устранить принципиально. В цифровых же системах связи принятая информация в большинстве случаев не содержит ошибок, то есть приёмник располагает оригинальными значениями битов — нулей и единиц.

Следует отметить, что теоретическое рассмотрение сигналов подразумевает их деление на типы: дискретный и непрерывный (причём как по времени, так и по уровню), однако, в рамках дисциплины «Цифровая связь» достаточно рассматривать сигналы как только непрерывные сигналы, что противопоставляет их дискретной по своей природе цифровой информации. Было бы глупо цифровую информацию рассматривать как некоторую непрерывную величину, но с другой стороны это и есть цифровая связь как таковая, потому что нельзя передавать информацию, не превратив её в некоторый сигнал, пригодный для передачи по заданной линии связи. Превращение информации в сигнал является насыщением информации энергией, что подчёркивает дуализм энергии и информации.

# Сигналы цифровых систем связи

## Энергия на один бит информации. Битовая скорость передачи информации.

Как теоретически, так и практически, в простейшем случае требуется передавать биты — двоичные символы. На передачу любой информации тратится некоторое время. Чем меньше тратится времени, тем выше скорость передачи информации. Сигналы цифровых систем связи характеризуются средней мощностью, которая вкупе с затраченным временем определяет затраченную энергию.

Для целей сравнения разных систем цифровых связи удобно рассматривать энергию на один бит информации:

$$E_b = P T_b.$$

Здесь $P$ — средняя мощность сигнала, $T_b$ — интервал следования битов (в секундах). При этом битовая скорость передачи информации будет равна:

$$V_b = 1 / T_b, ~бит/с.$$

Качество цифровых систем передачи информации напрямую зависит от энергии на один бит. Поясним почему это так.

## Почему качество цифровой связи определяется энергией, а не мощностью? Помехоустойчивое кодирование.

Само собой разумеется, что на качество связи влияет уровень шума/помех, однако, на данном этапе можно полагать, что этот уровень постоянен. Средняя мощность сигнала определяет «громкость» сигнала, которая хотя и влияет на качество связи, но не она одна. Длительность бита также влияет, потому что чем дольше происходит наблюдение некоторого сигнала, несущего один бит, тем потенциально больше накапливается промежуточной информации и тем выше вероятность правильного решения относительно переданного бита. Это тем актуальнее, чем ниже отношение сигнал-шум, т.е. хуже условия приёма. При отличных условиях приёма длительность бита можно уменьшить, сохранив то же качество связи, при этом повысится скорость передачи информации; справедливо и обратное: с ухудшением условий приёма увеличивают длительность информационного бита, чтобы сохранить требуемое качество связи, однако, скорость передачи при этом уменьшается. Существует некоторый предел для отношения сигнал-шум, при котором никакое увеличение длительности бита не восстановит исходного качества связи и произойдёт, фактически, обрыв связи (логический).

Заметим, что изменение длительности информационного бита можно понимать двояко:

1. Как физическое изменение интервала следования битов,
2. Как передачу информационного бита несколькими битами исходной длительности.

В первом случае изменяется ширина спектра сигнала; во втором случае происходит избыточное кодирование информации (помехоустойчивое кодирование), когда, допустим, $k$ исходных битов передают с помощью $n > k$ битов. На практике чаще применяют второй способ поддержания заданного качества связи.

## Интенсивный и экстенсивный пути улучшения качества цифровой связи.

Так как качество цифровых систем передачи информации напрямую зависит от энергии на один бит, то намечаются два пути улучшения качества связи:

1. Увеличение средней мощности сигнала,
2. Уменьшение скорости передачи информации (увеличение времени передачи информационного бита).

Первый путь является экстенсивным в том плане, что он эквивалентен, например, расширению площади засеваемых земель для повышения объема заготавливаемой продукции. Очевидно, существует интенсивный путь, который в плане сельского хозяйства эквивалентен повышению урожайности культуры за счет разного рода селекций (генетических манипуляций). Интенсивный путь улучшения качества цифровых систем связи — использование помехоустойчивых кодов, которые, естественно, не бесплатные, однако, затраты в большей степени связаны с усложнением процедуры кодирования информации, а также усложнением сигнального представления. Это приводит к определенной задержке обработки информации, а также увеличению потребляемой системой мощности в целом, хотя это увеличение относительно небольшое, если сравнивать с затратами экстенсивного пути.

Второй путь, если его понимать как изменение интервала следования битов, также экстенсивен, потому что время — это такой же количественный ресурс как и средняя мощность сигнала.

Несомненно, интенсивным путём будет улучшение общих технологий передачи/приёма сигналов. Сюда относятся:

* Уменьшение уровня шума приемников (транзисторные и другие технологии),
* Улучшение характеристик антенных систем (технологии материалов и поверхностей),
* Повышение стабильности генерируемых колебаний (технологии резонаторов),
* Повышение КПД передатчиков сигналов (транзисторные технологии),
* Улучшение экранирования подсистем (технологии материалов и поверхностей).

Однако с точки зрения общей теории связи все улучшения в итоге сводятся к:

* Уменьшению вероятности ошибки приема информационного бита,
* Уменьшению затрат энергии для передачи/приема информационного бита,
* Увеличению скорости передачи/приема информационного бита.

Следует отметить, что в реальных системах связи применяются все пути, несмотря на их экстенсивность/интенсивность. Это связано с требованием адаптации систем связи, потому что каналы связи могут быть достаточно изменчивы по своей природе (например, радиоканалы), а также есть потребность регулирования параметров по разным протокольным обстоятельствам, которые не зависят от качества канала связи — это разного рода регуляторные ограничения, например, на уровень излучения, на ширину канала.

С точки зрения системы «в целом», введение экстенсивных мер сглаживается наличием интенсивных, то есть последние, фактически, обеспечивают поддержание приемлемой эффективности системы при любых допустимых регулировках. Именно поэтому любые интенсивные меры — коды, алгоритмы, сигналы, технологии — очень ценны.

## Техническая и битовая скорости передачи. Символ модуляции. Порядок модуляции.

В цифровых системах передачи информации следует различать две скорости передачи информации — техническую и битовую.

Техническая скорость передачи информации определяется интервалом следования символов, $T$, при этом этот интервал прямым образом определяет длительность импульса, и не важно что данный импульс переносит: служебный ли бит, информационный, а может и вообще не бит, а целое кодовое слово — символ модуляции (несколько бит):

$$V_T = 1/T, ~бод.$$

Техническая скорость измеряется в бодах. Именно она главным образом определяет ширину канала (в герцах), который требуется для передачи данного сигнала. Например, интервалу следования символов 1 мс (1000 символов в секунду) соответствует скорость 1000 бод или 1 Кбод, при этом ширина канала должна быть порядка 1 кГц. Точное определение ширины канала возможно только при конкретизации системы связи, когда четко определен её состав.

Использование помехоустойчивых кодов приводит к уменьшению скорости передачи информации, потому что требуется успевать передавать служебные (проверочные) биты. Помехоустойчивые коды характеризуются так называемой избыточностью — количеством проверочных битов $r$ по отношению к общей длине кодового слова $n$:

$$R = r / n, ~0 \leq R \leq 1.$$

Битовая скорость — это скорость передачи информационных битов, не считая разного рода служебных битов, поэтому в текущем контексте она может быть определена так

$$V_b = 1 / T_b = (1 - R) V_T.$$

Нулевой избыточности соответствует отсутствие помехоустойчивого кодирования, поэтому битовая скорость совпадает с технической (считаем при этом, что никаких других служебных битов нет). Избыточности $0,5$ соответствует ситуация, когда один информационный бит передаётся двумя битами, т.е. битовая скорость в два раза ниже технической.

Чтобы не было драматического уменьшения битовой скорости от введения кодирования, вводят модуляцию, когда один символ модуляции представляет слово из нескольких битов. В этом случае выражение битовой скорости следует уточнить:

$$V_b = (1 - R) V_T \log_2 M, ~бит/с.$$

Здесь $M$ — количество символов модуляции (порядок модуляции, размер алфавита). В случае двоичной модуляции символов два, $M = 2$, и один символ переносит ровно один бит. В случае четверичной модуляции символов четыре, $M = 4$, и один символ переносит два бита, и т.д.

## О размерности энергии. Децибел.

Отметим, что хотя энергия $E$ и измеряется в джоулях (как абсолютная физическая величина), в практике цифровой связи используют безразмерную амплитуду сигналов, поэтому остаётся привязка только к секунде, и следовало бы энергии приписывать размерность $[с]$, однако, это сбивало бы с толку, поэтому пишут $[В^2 c]$, где $В$ — условный «вольт», который по факту безразмерен.

При сравнении систем связи друг с другом строят графики зависимости вероятности битовой ошибки от отношения сигнал-шум (ОСШ), при этом под отношением сигнал-шум понимают безразмерную величину

$$q^2 = E_b / N_0,$$

которая равна отношению средней энергии на один бит к спектральной плотности мощности (СПМ) шума. Зачастую ОСШ — энергетическую величину — удобно выражать в децибелах:

$${q}_{\text{дБ}} = 10 \lg E_b / N_0, ~дБ.$$

## Средняя мощность сигнала. Первичное определение.

Как уже было условлено, под сигналом будем понимать непрерывный сигнал, поэтому его можно обозначить как функцию времени, $s(t)$. Сигналы систем связи — это функции, которые возвращают комплексное значение. Напомним, что в системах связи величину сигнала (т.е. его амплитуду) принято считать безразмерной.

Мгновенной мощностью называется квадрат амплитуды сигнала, $p(t) = {\lvert s(t) \rvert}^2$. Тогда средней мощностью на интервале времени $T$ будет величина:

$$P = 1/T \int_{t_0}^{t_0 + T} {p(t) dt}.$$

Здесь происходит суммирование мгновенных мощностей на некотором интервале времени и деление суммы на этот интервал. Моментом начала суммирования является произвольно выбранный момент $t_0$.

Для сигналов цифровых систем связи характерна своего рода стационарность, когда сигнал — если наблюдать десятки и сотни символов модуляции подряд — выглядит достаточно однообразным, поэтому есть основания предполагать стабилизацию средней мощности $P$ при увеличении интервала измерения $T$, а также её независимость от момента начала измерения.

## Средняя мощность цифрового сигнала.

Да, это не опечатка. Снова про среднюю мощность, но уже с другой — более определённой стороны. Введённая выше средняя мощность обладает тем недостатком, что она приближённая, т.к. зависит от двух произвольных параметров. Для систем же связи можно выделить характерный момент времени — интервал следования символов модуляции, который определяет техническую скорость передачи информации (в бодах). Поэтому появляется возможность задать интервал интегрирования равным (или кратным) символьному интервалу. Однако, при фиксации интервала наблюдения точность измерения падает (точнее, не контролируется), поэтому усреднение, в итоге, делают статистически:

$$P = {{1} \over {kT}} \overline{\int_{t_0}^{t_0 + kT} {p(t) dt}}.$$

Под статистическим усреднением в контексте цифровых систем связи понимают усреднение по всем информационным импульсам, которые соответствуют символам модуляции:

$$P = {{1} \over {kT}} \sum_{i=0}^{M-1} \int_{t_0}^{t_0 + kT} {p_i {\lvert s_i(t) \rvert}^2 dt}.$$

Здесь $p_i$ — вероятность $i$-го символа модуляции, которому соответствует импульс $s_i(t)$.

Обычно, все символы модуляции равновероятные. Параметр $k = 1, 2, ...$ определяется длительностью информационного импульса, который был выбран для сигнального представления символов модуляции. Импульс сам по себе может занимать несколько символьных интервалов — это особенно актуально для систем радиосвязи, в которых важна компактность спектра излучаемого радиосигнала.

Определённая таким образом мощность является в каком-то смысле прецизионной — самой точной.

## Средняя мощность сигнала по сигнальному созвездию.

И, наконец, последняя ремарка про среднюю мощность. На этот раз с самой простой и практичной стороны.

Недостатком выше определённой средней мощности является сложность расчётной формулы и зависимость от стороннего параметра $k$, который в большей степени определяет компактность спектра формируемого сигнала, чем среднюю мощность. Чтобы упростить формулу расчёта средней мощности, интеграл заменяют конкретным значением, а суммирование оставляют неизменным:

$$P = \sum_{i=0}^{M-1} {p_i {\lvert s_i \rvert}^2}.$$

Конкретное значение ${\lvert s_i \rvert} \neq 0$ выбирают равным кодовому числу, которое соответствует $i$-му символу модуляции. Таким числом является комплексное число, которое отображается в виде точки на комплексной плоскости, при этом набор таких точек называют сигнальным созвездием, которое и задаёт вид модуляции.

## Цифровая модуляция (манипуляция). Модель цифрового сигнала.

Процесс преобразования символа модуляции $s_i$ в некоторый сигнал $s_i(t)$ называется цифровой модуляцией или просто манипуляцией. Обратный процесс, соответственно, демодуляцией. Модуляция соответствует передающей стороне (передатчику), а демодуляция — приёмной стороне (приёмнику).

Из-за неидеальности канала связи процесс демодуляции сопровождается ошибками. Качество демодуляции определяется вероятностью символьной ошибки (частотой символьных ошибок), которая зависит от отношения сигнал-шум в приёмнике. Каждый вид модуляции имеет свою зависимость вероятности символьной ошибки от отношения сигнал-шум. Вероятность символьной ошибки можно — точно или приближенно — пересчитать в вероятность ошибки на один бит, чтобы была возможность сравнивать разные системы связи друг с другом.

Модуляция (манипуляция) бывает:

* Линейной, когда информационный импульс $g(t)$ прямо пропорционально модулирует сигнал $s(t)$:

$$v_I(t) = \sum_{n \in \mathbb{Z}} {I_n g(t - nT)}, ~~v_Q(t) = \sum_{n \in \mathbb{Z}} {Q_n g(t - nT)},$$

$$s(t) = \left[v_I(t) + j v_Q(t)\right] e^{j 2 \pi f_0 t}.$$

* Нелинейной, когда информационный импульс $g(t)$ модулирует сигнал $s(t)$ через некоторое нелинейное преобразование $F(v(t))$, например:

$$s(t) = A e^{j \theta(t)} = A e^{j 2 \pi f_0 t + K \int_{-\infty}^{t} {j v(t)} dt}.$$

Здесь для случая нелинейной модуляции приведён пример частотной манипуляции (ЧМн), когда мгновенная частота сигнала $s(t)$ управляется сигналом $v(t)$:

$$f(t) := {1/2\pi} {{d \theta} \over {dt}} = {1/2\pi} {{d} \over {dt}} \left[ {2 \pi f_0 t + K \int_{-\infty}^{t} {v(t)} dt} \right] = {f_0 + {{K} \over {2\pi}} v(t)}.$$

Под $I_n$ и $Q_n$ понимаются символы, т.е. непосредственно цифровой поток в дискретном времени $n \in \mathbb{Z}$. Символы как правило принимают нечётные значения из диапазона целых чисел $\pm 1, \pm 3, ...$, хотя в общем случае цифровая последовательность принимает произвольные целые числа, зависящие от конкретной схемы модуляции.

Буквы $I$ и $Q$ обозначают квадратурные компоненты - проекции текущей точки на комплексной плоскости на оси $OX$ и $OY$, соответственно.

Под $f_0$ понимают некоторую несущую частоту (либо промежуточную частоту), которой в общем случае может и не быть (например, в проводных системах связи с медным кабелем).

Под $\theta(t)$ понимают полную фазу сигнала согласно модели сигнала в виде зависящей от времени комплексной амплитуды.

Для случая линейной модуляции записана общая модель цифрового сигнала для амплитудной (АМн), фазовой (ФМн), амплитудно-фазовой (АФМн) и квадратурной манипуляций (КАМ-М).

## Информационные импульсы. Тактовая синхронизация.

Перед тем как передавать цифровую информацию через канал (линию) связи, требуется определиться с сигналом, который будет переносить информацию.

В первом приближении требуемая ширина канала определяется интервалом следования символов модуляции, $T$:

$$\Delta f \sim 1/T, ~Гц.$$

Однако, длительность сигнала — информационного импульса $g(t)$

$${T}_{\text{pulse}} = kT, ~k=1,2,...,$$

и его форма также задают требования к каналу, но уже вторично, хотя и достаточно значимо, потому что качество систем связи определяется в том числе системой тактовой синхронизации, которая восстанавливает интервал следования символов модуляции с точностью до фазы. Слишком узкий канал приведёт к недопустимому искажению сигнала и сбою системы синхронизации. Система тактовой синхронизации настраивается исходя из особенностей формы сигнала и поэтому чувствительна к форме принимаемого сигнала.

## Формирующий и согласованный фильтры. Межсимвольная интерференция. Оптимальный прием.

Чем компактнее должен быть спектр формируемого цифрового сигнала, тем больше должна быть длительность информационного импульса. Это можно наглядно подкрепить той аналогией, что самый узкий по спектру сигнал «чистый тон» (синус) имеет бесконечную длительность. Практически же используют функции типа синуса Котельникова с разными оконными функциями (например, фильтр "приподнятого косинуса"). При этом чтобы исключить взаимное влияние импульсов в моменты приёма, от информационных импульсов $g(t)$ в простейшем случае требуется соблюдение отсчётного свойства:

$$g(nT) = 0, ~n \neq 0,$$
$$g(nT) = 1, ~n = 0.$$

Таким образом, в идеальном случае (при отсутствии шумов и при точной работе системы тактовой синхронизации) в точечные моменты времени кратно интервалу следования символов $T$ исключается межсимвольная интерференция.

Ниже проиллюстрировано отсчётное свойство фильтра "приподнятого косинуса" на примере приёма четырёх двоичных символов $[1, 1, -1, -1]$.

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

def raised_cosine(t, T, beta):
    """Функция приподнятого косинуса."""
    if beta == 0:
        return np.sinc(t / T)
    with np.errstate(divide='ignore', invalid='ignore'):
        # numpy.sinc(x) это sin(pi*x)/(pi*x)
        sinc = np.sinc(t / T)
        cos_part = np.cos(np.pi * beta * t / T)
        denom = 1 - (2 * beta * t / T)**2
        res = sinc * cos_part / denom
        # Обработка точек, где знаменатель равен 0
        res[np.abs(denom) < 1e-10] = (np.pi / 4) * np.sinc(1 / (2 * beta))
    return res

# 1. Создаем виджет
float_slider = widgets.FloatSlider(
    value=0.5,           # Начальное значение
    min=0,               # Минимум
    max=1.0,            # Максимум
    step=0.05,            # Шаг изменения
    description='Roll-off фактор:',
    # 'initial' заставит виджет подстроиться под длину текста
    style={'description_width': 'initial'},
    readout_format='.2f' # Формат отображения (1 знак после запятой)
)

# 2. Функция отрисовки
def plot_func(beta):
    # Параметры
    T = 1.0          # Символьный интервал
    sps = 12         # Отсчетов на символ
    span = 8        # Увеличим охват символов для красоты
    
    # Временная ось (расширена для трех импульсов)
    t = np.linspace(-span*T/2, span*T/2 + T, span*sps + sps + 1)
    
    # Генерация трех импульсов с шагом T
    p1 = 1*raised_cosine(t + T, T, beta) # Центрирован в -T
    p2 = 1*raised_cosine(t, T, beta)     # Центрирован в 0
    p3 = -1*raised_cosine(t - T, T, beta) # Центрирован в +T
    p4 = -1*raised_cosine(t - 2*T, T, beta) # Центрирован в +2T
    
    # Суммарный сигнал
    v = p1 + p2 + p3 + p4
    
    # Визуализация
    plt.figure(figsize=(10, 6))
    
    plt.plot(t, p1, 'g--', label='Импульс g(t+T)', alpha=0.5)
    plt.plot(t, p2, 'b--',  label='Импульс g(t)', alpha=0.5)
    plt.plot(t, p3, 'r--', label='Импульс g(t-T)', alpha=0.5)
    plt.plot(t, p4, 'm--', label='Импульс g(t-2T)', alpha=0.5)
    plt.plot(t, v, 'k', label='Суммарный сигнал v(t)', linewidth=1.7)
    
    # Моменты дискретизации
    sampling_instants = [-T, 0, T, 2*T]
    plt.scatter(sampling_instants, [1, 1, -1, -1], color='red', zorder=5, label='Точки отсчета')
    
    # Линии для наглядности
    for ts in sampling_instants:
        plt.axvline(ts, color='gray', linestyle=':', alpha=0.4)
    
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
    plt.title(f'Прием последовательности символов [+1, +1, -1, -1]', fontsize=19)
    plt.xlabel('Время (t/T)', fontsize=16)
    plt.ylabel('Амплитуда', fontsize=16)
    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.3)
    plt.axhline(0, color='black', linewidth=0.8)
    plt.ylim(-1.5, 1.5)
    
    plt.show()

# 3. Связываем виджет и функцию
# Ключ в словаре должен совпадать с названием аргумента функции
out = widgets.interactive_output(plot_func, {'beta': float_slider})

# 4. Отображаем всё вместе
display(float_slider, out)
```

Так как система связи предполагает две неотъемлемые части: передатчик (Tx) и приёмник (Rx), то исходный фильтр расщепляют на два идентичных фильтра: формирующий (передающий) и согласованный (приемный). Расщеплению фильтров в частотной области соответствует их каскадное соединение:

$$H(f) = \sqrt{H_\text{Tx}(f)} \sqrt{H_\text{Rx}(f)}.$$

Это позволяет в приёмнике естественным образом выполнять оптимальный приём, получая наибольшее отношение сигнал-шум после обработки:

$$q_\text{max}^2 = {{2E_d} \over {N_0}}.$$

Здесь под $E_d$ имеется ввиду энергия разностного сигнала, которую в каждом конкретном случае требуется пересчитывать в энергию на один бит $E_b$, чтобы можно было корректно сравнивать системы связи друг с другом. Под разностным сигналом понимается разность двух эталонных сигналов, амплитуда которых приведена к текущему уровню сигнала в приёмнике:

$$s_d(t) = s_i(t) - s_j(t).$$

Здесь $i \neq j$ обозначают две ветви в момент принятия решения. Для двоичных модуляций существуют всего два импульса, поэтому там решение принимается единожды. Для модуляций более высоких порядков решение принимается поэтапно. Например, для модуляции ФМн-4 сначала определяется правдоподобие для левой и правой полуплоскостей (анализируется только $I$ квадратура), а затем правдоподобие для верхней и нижней (при условии, что решение "левая-правая" уже принято). Таким образом, за два этапа угадывается четверичный символ, который по сути эквивалентен номеру четверти (квадранту комплексной плоскости). Для каждого этапа будет в общем случае своё отношение сигнал-шум.

Примером расщеплённого фильтра является фильтр типа "корень из приподнятого косинуса".

## Спектральная эффективность систем связи.

Ширина канала (двусторонняя) при использовании формирующего фильтра типа "приподнятый косинус" равна:

$$\Delta f = (1 + \beta) / T.$$

Здесь $~0 \leq\beta \leq 1~$ — коэффициент, определяемый формирующим фильтром (roll-off фактор).

Используя ширину канала и битовую скорость передачи, вводят понятие спектральной эффективности:

$$\eta_s = V_b / \Delta f = {{(1 - R) \log_2 M} \over {1 + \beta}}, ~бит/с/Гц.$$

Для систем мобильной связи типовые значения спектральной эффективности лежат в диапазоне $~0,1 < \eta_s < 10~$ в зависимости от стандарта: 2G, 3G, 4G и 5G. Перевал за единицу произошёл на стыке сетей третьего и четвертого поколений. Следует отметить, что эффективность систем связи - статистическая величина, т.к. современные системы связи адаптируются к текущим условиям приёма, переключая кодирование/модуляцию для поддержания заданной битовой вероятности ошибки.

## Энергетическая эффективность систем связи.

Данная величина является оборотной стороной медали по отношению к рассмотренной выше спектральной эффективности:

$$\eta_e = V_b / P, ~бит/Дж.$$

Здесь $P$ – средняя потребляемая системой мощность.

В контексте общей теории связи под средней мощностью понимают среднюю мощность, определяемую по сигнальному созвездию. Такая средняя мощность пропорциональна порядку модуляции, т.е. количеству символов созвездия:

$$P \sim M {d}^2_{\text {min}}$$

Здесь ${d}_{\text{min}}$ — минимальное расстояние между точками сигнального созвездия. Рост порядка модуляции улучшает спектральную эффективность, но ухудшает энергетическую.


## Ресурсная эффективность систем связи.

Взвешенное произведение спектральной и энергетической эффективностей дает ресурсную эффективность системы связи:

$$\eta_r = {\eta_s}^{w} {\eta_e}^{1 - w}$$

В зависимости от типа системы связи отдают приоритет либо спектральной эффективности, либо энергетической, регулируя параметр $0 \leq w \leq 1$. Современные мобильные системы связи (5G и 6G) могут регулировать этот параметр, например, в зависимости от времени суток: днём отдавая приоритет спектральной эффективности, а ночью — энергетической.

Иногда говорят об энерго-спектральной эффективности, когда при проектировании находят «точку равновесия», при которой система не тратит лишнюю энергию ради незначительного прироста скорости.
