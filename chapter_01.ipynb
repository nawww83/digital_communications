# Конспекты по «Цифровой связи»

Представлено конспективное изложение основных моментов «Цифровой связи» — дисциплины, которая может скрываться под разными названиями, такими как «Общая теория связи», «Передача дискретных сообщений», «Теория и техника передачи информации», «Формирование и обработка сигналов систем связи» и т. п.

Целью поставлено изложение материала в достаточно общей форме без глубокого погружения в детали, что даст изучающему фундамент, который позволит изучать более детальные (но и менее доступные для понимания) книги, учебные пособия и статьи.

Данное изложение является результатом переосмысления опыта 9-летнего преподавания данной дисциплины в техническом ВУЗе.

Материал изложен в виде относительно небольшого количества абзацев, разделенных подзаголовками, что позволяет быстро отыскивать материал в случае надобности, а также сосредотачиваться на так называемых «моментах» и осмысливать их дольше, нежели они были бы обычными абзацами достаточно большой главы.

Базовыми требованиями являются знание высшей математики и теории вероятностей технического вуза, а также школьной физики. Будет плюсом знание дисциплины «Радиотехнические цепи и сигналы» и/или «Основы теории цепей».

# Введение

## Цифровая связь как транспортный сервис.

Под цифровой связью подразумевают техническую передачу символов с помощью некоторых сигналов через определенные каналы связи. Символами называют элементы заранее определенного алфавита, при этом алфавитами не являются алфавиты в традиционном понимании, — алфавитами являются таблицы соответствия между двоичными словами (как правило фиксированной длины) и сигналами определенной формы — импульсами. Именно эта формальность подчеркивает техническую сторону цифровой связи как некоторого транспортного сервиса, которому безразлично содержание передаваемой информации.

## Канал связи. Качество канала связи. Надежная связь.

Под каналом связи понимают согласованный набор технических средств, который позволяет, задавая определенное входное воздействие, наблюдать связанный с ним отклик — выходное воздействие. Чем лучше мы можем судить о входном воздействии по наблюдаемому отклику, тем лучше качество канала связи. При этом для поддержания заданного качества в течение заданного времени тратится некоторая энергия. Чем выше требуемое качество, тем больше тратится энергии, однако, основная соль в том, что кривая зависимости затрачиваемой энергии от качества зависит от конкретного канала связи; стремятся, чтобы рост качества не сопровождался бы драматическим ростом потребляемой энергии.

Качество канала связи в цифровых системах передачи информации количественно определяется вероятностью ошибки наблюдаемого (принимаемого) символа. Чем качественнее канал, тем меньше повторных запросов требуется для обеспечения надежной связи, при этом под надежной связью понимают практически идеальную передачу информации, когда итоговой вероятностью ошибки можно пренебречь в рамках решаемой задачи.

## Линия связи.

Под линией связи понимают некоторую физическую среду передачи, которая направляет сигнал, несущий передаваемую информацию. Направление необходимо для экономии энергии и для разграничения связи по каналам, чтобы не мешать параллельно существующим системам связи. Примерами направляющих сред могут быть любые кабели (телефонный кабель типа «лапша», витая пара, коаксиальный кабель, оптическое волокно), волноводы, водная и воздушная среды. В случае водной и воздушной среды направленность сама по себе слабая, однако, она создается антенными системами, которые, в принципе, можно отнести к направляющим средам. Линия связи входит в канал связи (вложена в него) и не может быть из него исключена принципиально.

## Непрерывность сигнала, дискретность цифровой информации.

Под сигналом понимают некоторую величину, которая изменяется с течением времени. В системах связи часто это напряжение или ток в электрической цепи, интенсивность света в оптическом волокне, комплексная амплитуда электромагнитного СВЧ поля в линии. При этом этой изменяющейся величине соответствует некоторая информация.

В системах цифровой связи информация является дискретной по своей природе (т. е. счётной); сам же сигнал всегда непрерывен по своей природе безотносительно к типу рассматриваемой системы. В аналоговых системах связи информацией фактически является непрерывный сигнал, полученный после всех преобразований в приемнике; неотъемлемой частью такого сигнала является шум, который можно снижать, но нельзя устранить принципиально. В цифровых же системах связи принятая информация в большинстве случаев не содержит ошибок, то есть приемник располагает оригинальными значениями битов — нулей и единиц.

Следует отметить, что теоретическое рассмотрение сигналов подразумевает их деление на типы: дискретный и непрерывный (причем как по времени, так и по уровню), однако, в рамках дисциплины «Цифровая связь» достаточно рассматривать сигналы как только непрерывные сигналы, что противопоставляет их дискретной по своей природе цифровой информации. Было бы глупо цифровую информацию рассматривать как некоторую непрерывную величину, но с другой стороны это и есть цифровая связь как таковая, потому что нельзя передавать информацию, не превратив ее в некоторый сигнал, пригодный для передачи по заданной линии связи.

# Сигналы цифровых систем связи

## Энергия на один бит информации. Битовая скорость передачи информации.

Как теоретически, так и практически, в простейшем случае требуется передавать биты — двоичные символы. На передачу любой информации тратится некоторое время. Чем меньше тратится времени, тем выше скорость передачи информации. Сигналы цифровых систем связи характеризуются средней мощностью, которая вкупе с затраченным временем определяет затраченную энергию.

Для целей сравнения разных систем цифровых связи удобно рассматривать энергию на один бит информации:

$$E_b = P T_b.$$

Здесь $P$ — средняя мощность сигнала, $T_b$ — интервал следования битов (в секундах). При этом битовая скорость передачи информации будет равна:

$$V_b = 1 / T_b, ~бит/с.$$

Качество цифровых систем передачи информации напрямую зависит от выше рассмотренной энергии на один бит. Поясним почему это так.

## Почему качество цифровой связи определяется энергией, а не мощностью? Помехоустойчивое кодирование.

Само собой разумеется, что на качество связи влияет уровень шума/помех, однако, на данном этапе можно полагать, что этот уровень постоянен. Средняя мощность сигнала определяет «громкость» сигнала, которая хотя и влияет на качество связи, но не она одна. Длительность бита также влияет, потому что чем дольше происходит наблюдение некоторого сигнала, несущего один бит, тем потенциально больше накапливается промежуточной информации и тем выше вероятность правильного решения относительно переданного бита. Это тем актуальнее, чем ниже отношение сигнал-шум. При отличных условиях приема длительность бита можно уменьшить, сохранив то же качество связи, при этом повысится скорость передачи информации; справедливо и обратное: с ухудшением условий приема увеличивают длительность информационного бита, чтобы сохранить требуемое качество связи, однако, скорость передачи при этом уменьшается. Существует некоторый предел для отношения сигнал-шум, при котором никакое увеличение длительности бита не восстановит исходного качества связи и произойдет, фактически, обрыв связи (логический).

Заметим, что изменение длительности информационного бита можно понимать двояко:

1. Как физическое изменение интервала следования битов,
2. Как передачу информационного бита несколькими битами исходной длительности.

В первом случае изменяется ширина спектра сигнала; во втором случае происходит избыточное кодирование информации (помехоустойчивое кодирование), когда, допустим, 27 исходных битов передают с помощью 32 битов, и т.п. На практике чаще применяют второй способ поддержания заданного качества связи.

## Интенсивный и экстенсивный пути улучшения качества цифровой связи.

Так как качество цифровых систем передачи информации напрямую зависит от энергии на один бит, то намечаются два пути улучшения качества связи:

1. Увеличение средней мощности сигнала,
2. Уменьшение скорости передачи информации (увеличение времени передачи информационного бита).

Первый путь является экстенсивным в том плане, что он эквивалентен, например, расширению площади засеваемых земель для повышения объема заготавливаемой продукции. Очевидно, существует интенсивный путь, который в плане сельского хозяйства эквивалентен повышению урожайности культуры за счет разного рода селекций (генетических манипуляций). Интенсивный путь улучшения качества цифровых систем связи — использование помехоустойчивых кодов, которые, естественно, не бесплатные, однако, затраты в большей степени связаны с усложнением процедуры кодирования информации, а также усложнением сигнального представления. Это приводит к определенной задержке обработки информации, а также увеличению потребляемой системой мощности в целом, хотя это увеличение относительно небольшое, если сравнивать с затратами экстенсивного пути.

Второй путь, если его понимать как изменение интервала следования битов, также экстенсивен, потому что время — это такой же количественный ресурс как и средняя мощность сигнала. Чем короче сигнал, тем меньше времени тратится на передачу одного бита и, следовательно, выше скорость передачи. Очевидно, что высокая скорость передачи может привести к ненадежному приему по вполне обычной аналогии, например, с человеческой речью.

Несомненно, интенсивным путем будет улучшение общих технологий передачи/приема сигналов. Сюда относятся:

* Уменьшение уровня шума приемников (транзисторные и другие технологии),
* Улучшение характеристик антенных систем (технологии материалов и поверхностей),
* Повышение стабильности генерируемых колебаний (технологии резонаторов),
* Повышение КПД передатчиков сигналов (транзисторные технологии),
* Улучшение экранирования подсистем (технологии материалов и поверхностей).

Однако с точки зрения общей теории связи все улучшения в итоге сводятся к:

* Уменьшению вероятности ошибки приема информационного бита,
* Уменьшению затрат энергии для передачи/приема информационного бита,
* Увеличению скорости передачи/приема информационного бита.

Следует отметить, что в реальных системах связи применяются все пути, несмотря на их экстенсивность/интенсивность. Это связано с требованием адаптации систем связи, потому что каналы связи могут быть достаточно изменчивы по своей природе (например, радиоканалы), а также есть потребность регулирования скоростей передачи по разным протокольным обстоятельствам, которые не зависят от качества канала связи — это разного рода регуляторные ограничения, например, на уровень излучения, на ширину канала.
С точки зрения системы «в целом», введение экстенсивных мер сглаживается наличием интенсивных, то есть последние, фактически, обеспечивают поддержание эффективности системы при любых допустимых регулировках. Именно поэтому любые интенсивные меры — коды, алгоритмы, сигналы, технологии — очень ценны.

## Техническая и битовая скорости передачи. Символ модуляции. Порядок модуляции.

В цифровых системах передачи информации следует различать две скорости передачи информации — техническую и битовую.

Техническая скорость передачи информации определяется интервалом следования символов, $T$, при этом этот интервал прямым образом определяет длительность импульса, и не важно что данный импульс переносит: служебный ли бит, информационный, а может и вообще не бит, а целое кодовое слово — символ модуляции (несколько бит):

$$V_T = 1/T, ~бод.$$

Техническая скорость измеряется в бодах. Именно она главным образом определяет ширину канала (в герцах), который требуется для передачи данного сигнала. Например, интервалу следования символов 1 мс (1000 символов в секунду) соответствует скорость 1000 бод или 1 Кбод, при этом ширина канала должна быть порядка 1 кГц. Точное определение ширины канала возможно только при конкретизации системы связи, когда четко определен её состав.

Использование помехоустойчивых кодов приводит к уменьшению скорости передачи информации, потому что требуется успевать передавать служебные (проверочные) биты. Помехоустойчивые коды характеризуются так называемой избыточностью — количеством проверочных битов $r$ по отношению к общей длине кодового слова $n$ в битах:

$$R = r / n, ~0 \leq R \leq 1.$$

Битовая скорость — это скорость передачи информационных битов, не считая разного рода служебных битов, поэтому в текущем контексте она может быть определена так

$$V_b = 1 / T_b = (1 - R) V_T.$$

Нулевой избыточности соответствует отсутствие помехоустойчивого кодирования, поэтому битовая скорость совпадает с технической (считаем при этом, что никаких других служебных битов нет)

Чтобы не было драматического уменьшения скорости от введения кодирования (которое может достигать 50% и выше), вводят модуляцию, так, что один символ модуляции кодирует несколько битов. В этом случае формулу для битовой скорости следует уточнить:

$$V_b = (1 - R) V_T \log_2 M, ~бит/с.$$

Здесь $M$ — количество символов модуляции (размер алфавита, порядок модуляции). В случае двоичной модуляции символов два, $M = 2$, и один символ переносит ровно один бит. В случае четверичной модуляции символов четыре, $M = 4$ , и один символ переносит два бита, и т. д.

## О размерности энергии.

Отметим, что хотя энергия $E$ и измеряется в джоулях (как абсолютная физическая величина), в практике цифровой связи используют безразмерную амплитуду сигналов, поэтому остается привязка только к секунде, и следовало бы энергии приписывать размерность $[с]$, однако, это сбивало бы с толку, поэтому пишут $В^2 c$, где $В$ — условный «вольт», который по факту безразмерен.

При сравнении систем связи друг с другом строят графики зависимости вероятности битовой ошибки от отношения сигнал-шум (ОСШ), при этом под отношением сигнал-шум понимают безразмерную величину

$$q^2 = E_b / N_0,$$

которая равна отношению средней энергии на один бит к спектральной плотности мощности шума. Зачастую ОСШ — энергетическую величину — удобно выражать в децибелах:

$${q}_{\text{дБ}} = 10 \lg E_b / N_0, ~дБ.$$

## Средняя мощность сигнала. Первичное определение.

Как уже было условлено, под сигналом будем понимать непрерывный сигнал, поэтому его можно обозначить как функцию времени, $s(t)$. Сигналы систем связи — это функции, которые возвращают комплексное значение. Напомним, что в системах связи принято величину сигнала (т. е. его амплитуда) считать безразмерной величиной.

Мгновенной мощностью называется квадрат амплитуды сигнала, $p(t) = {\lvert s(t) \rvert}^2$ . Средней мощностью сигнала называют среднюю мощность на интервале времени $T$. Объективным аппаратом для вычисления только что определенной средней мощности является интеграл:

$$P = 1/T \int_{t_0}^{t_0 + T} {p(t) dt}.$$

Здесь по факту происходит суммирование мгновенных мощностей на некотором интервале времени и деление суммы на этот интервал. Моментом начала суммирования является произвольно выбранный момент $t_0$.

Для сигналов цифровых систем связи характерна своего рода монотонность, когда сигнал со стороны — если наблюдать десятки и сотни символов модуляции подряд — выглядит достаточно однообразным, поэтому есть основания предполагать стабилизацию средней мощности при увеличении интервала измерения $T$, а также ее независимость от момента начала измерения.

## Средняя мощность цифрового сигнала.

Да, это не опечатка. Снова про среднюю мощность, но уже с другой — более определенной стороны. Введенная выше средняя мощность обладает тем недостатком, что она приближенная, так как зависит от двух произвольных параметров. Для систем связи можно выделить характерный момент времени — интервал следования символов модуляции, который определяет техническую скорость передачи информации (в бодах). Поэтому появляется возможность задать интервал интегрирования (суммирования) равным (или кратным) символьному интервалу. Однако, при фиксации интервала наблюдения точность измерения падает (не контролируется), поэтому усреднение, в итоге, делают статистически:

$$P = {{1} \over {kT}} \overline{\int_{t_0}^{t_0 + kT} {p(t) dt}}.$$

Под статистическим усреднением в контексте цифровых систем связи понимают усреднение по всем информационным импульсам, которые соответствуют символам модуляции:

$$P = {{1} \over {kT}} \sum_{i=0}^{M-1} \int_{t_0}^{t_0 + kT} {p_i {\lvert s_i(t) \rvert}^2 dt}.$$

Здесь $p_i$ — вероятность $i$-го символа модуляции, которому соответствует импульс $s_i(t)$. Обычно все символы модуляции равновероятные. Параметр $k = 1, 2, ...$ определяется длительностью информационного импульса, который был выбран для сигнального представления символов модуляции. Импульс сам по себе может занимать несколько символьных интервалов — это актуально для систем радиосвязи, в которых важна компактность спектра излучаемого сигнала. Определенная таким образом мощность является в каком-то смысле прецизионной — самой точной.

## Средняя мощность сигнала по сигнальному созвездию.

И, наконец, последняя ремарка про среднюю мощность. На этот раз с самой простой и практичной стороны. Недостатком выше определенной средней мощности является сложность расчетной формулы и зависимость от стороннего параметра $k$, который в большей степени определяет компактность спектра формируемого сигнала, чем среднюю мощность.
Чтобы упростить формулу расчета средней мощности, интеграл заменяют конкретным значением, а суммирование оставляют неизменным:

$$P = \sum_{i=0}^{M-1} {p_i {\lvert s_i \rvert}^2}.$$

Конкретное значение ${\lvert s_i \rvert} \neq 0$ выбирают равным тому кодирующему числу, которое соответствует символу модуляции. Такому числу соответствует некоторое комплексное число, которое можно отобразить в виде точки на комплексной плоскости. Набор таких заранее заданных точек называют сигнальным созвездием, которое задаёт вид модуляции.

## Модуляция. Демодуляция.

Процесс преобразования символа модуляции $s_i$ в сигнал $s_i(t)$ называется цифровой модуляцией. Обратный процесс, соответственно, демодуляцией. Модуляция соответствует передающей стороне (передатчику), а демодуляция — приемной стороне (приемнику).

Из-за неидеальности канала связи процесс демодуляции сопровождается ошибками. Качество демодуляции определяется вероятностью символьной ошибки (частотой символьных ошибок), которая зависит от отношения сигнал-шум в приёмнике. Каждый вид модуляции имеет свою зависимость вероятности символьной ошибки от отношения сигнал-шум. Вероятность символьной ошибки можно — точно или приближенно — пересчитать в вероятность ошибки на один бит.

## Информационные импульсы. Тактовая синхронизация.

Перед тем как передавать цифровую информацию через канал (линию) связи, требуется определиться с сигналом, который будет переносить информацию.

В первом приближении требуемая ширина канала определяется интервалом следования символов модуляции, $T$:

$$\Delta f \sim 1/T, ~Гц.$$

Однако, длительность сигнала (информационного импульса)

$${T}_{\text{pulse}} = kT,$$

и его форма также задают требования к каналу, но уже вторично, хотя и достаточно значимо, потому что качество систем связи определяется в том числе системой тактовой синхронизации, которая восстанавливает интервал следования символов модуляции с точностью до фазы. Слишком узкий канал приведет к недопустимому искажению сигнала и сбою системы синхронизации. Система тактовой синхронизации настраивается исходя из особенностей формы сигнала и поэтому чувствительна к форме принимаемого сигнала.

## Формирующий и согласованный фильтры. Межсимвольная интерференция. Оптимальный прием.

Чем компактнее должен быть спектр формируемого цифрового сигнала, тем больше должна быть длительность информационного импульса. Это можно наглядно подкрепить той аналогией, что самый узкий по спектру сигнал — «чистый тон» (синус), который имеет бесконечную длительность. Практически используют функции типа синуса Котельникова с разными оконными функциями (например, фильтр "приподнятого косинуса"). При этом чтобы исключить взаимное влияние импульсов в моменты приема, от информационных импульсов $g(t)$ в простейшем случае требуется соблюдение отсчетного свойства:

$$g(nT) = 0, ~n \neq 0,$$
$$g(nT) = 1, ~n = 0.$$

— таким образом в идеальном случае (при отсутствии шумов и при точной работе системы тактовой синхронизации) в точечные моменты времени кратно интервалу следования символов $T$ исключается межсимвольная интерференция.

Ниже проиллюстрировано отсчетное свойство фильтра "приподнятого косинуса".

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

def raised_cosine(t, T, beta):
    """Функция приподнятого косинуса."""
    if beta == 0:
        return np.sinc(t / T)
    with np.errstate(divide='ignore', invalid='ignore'):
        # numpy.sinc(x) это sin(pi*x)/(pi*x)
        sinc = np.sinc(t / T)
        cos_part = np.cos(np.pi * beta * t / T)
        denom = 1 - (2 * beta * t / T)**2
        res = sinc * cos_part / denom
        # Обработка точек, где знаменатель равен 0
        res[np.abs(denom) < 1e-10] = (np.pi / 4) * np.sinc(1 / (2 * beta))
    return res

# 1. Создаем виджет
float_slider = widgets.FloatSlider(
    value=0.5,           # Начальное значение
    min=0,               # Минимум
    max=1.0,            # Максимум
    step=0.05,            # Шаг изменения
    description='Roll-off фактор:',
    # 'initial' заставит виджет подстроиться под длину текста
    style={'description_width': 'initial'},
    readout_format='.2f' # Формат отображения (1 знак после запятой)
)

# 2. Функция отрисовки
def plot_func(beta):
    # Параметры
    T = 1.0          # Символьный интервал
    sps = 12         # Отсчетов на символ
    span = 8        # Увеличим охват символов для красоты
    
    # Временная ось (расширена для трех импульсов)
    t = np.linspace(-span*T/2, span*T/2 + T, span*sps + sps + 1)
    
    # Генерация трех импульсов с шагом T
    p1 = 1*raised_cosine(t + T, T, beta) # Центрирован в -T
    p2 = 1*raised_cosine(t, T, beta)     # Центрирован в 0
    p3 = -1*raised_cosine(t - T, T, beta) # Центрирован в +T
    p4 = -1*raised_cosine(t - 2*T, T, beta) # Центрирован в +2T
    
    # Суммарный сигнал
    v = p1 + p2 + p3 + p4
    
    # Визуализация
    plt.figure(figsize=(10, 6))
    
    plt.plot(t, p1, 'g--', label='Импульс g(t+T)', alpha=0.5)
    plt.plot(t, p2, 'b--',  label='Импульс g(t)', alpha=0.5)
    plt.plot(t, p3, 'r--', label='Импульс g(t-T)', alpha=0.5)
    plt.plot(t, p4, 'm--', label='Импульс g(t-2T)', alpha=0.5)
    plt.plot(t, v, 'k', label='Суммарный сигнал v(t)', linewidth=1.7)
    
    # Моменты дискретизации
    sampling_instants = [-T, 0, T, 2*T]
    plt.scatter(sampling_instants, [1, 1, -1, -1], color='red', zorder=5, label='Точки отсчета')
    
    # Линии для наглядности
    for ts in sampling_instants:
        plt.axvline(ts, color='gray', linestyle=':', alpha=0.4)
    
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
    plt.title(f'Прием последовательности символов [+1, +1, -1, -1]', fontsize=19)
    plt.xlabel('Время (t/T)', fontsize=16)
    plt.ylabel('Амплитуда', fontsize=16)
    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.3)
    plt.axhline(0, color='black', linewidth=0.8)
    plt.ylim(-1.5, 1.5)
    
    plt.show()

# 3. Связываем виджет и функцию
# Ключ в словаре должен совпадать с названием аргумента функции
out = widgets.interactive_output(plot_func, {'beta': float_slider})

# 4. Отображаем всё вместе
display(float_slider, out)
```

Так как система связи предполагает две неотъемлемые части: передатчик и приемник, то исходный фильтр расщепляют на два идентичных, но имеющих разные названия, фильтра: формирующий (передающий) и согласованный (приемный). Расщепление происходит в частотной области. Это позволяет в приемнике выполнять оптимальный приём, получая наибольшее отношение сигнал-шум после обработки. Примером расщепленного фильтра является фильтр типа "корень из приподнятого косинуса".

## Спектральная эффективность систем связи.

Ширина канала (двусторонняя) при использовании формирующего фильтра равна:

$$\Delta f = (1 + \beta) / T.$$

Здесь $~0 \leq\beta \leq 1~$ — коэффициент, определяемый формирующим фильтром (roll-off фактор).

Используя ширину канала и битовую скорость передачи, вводят понятие спектральной эффективности:

$$\eta_s = V_b / \Delta f = {{(1 - R) \log_2 M} \over {1 + \beta}}, ~бит/с/Гц.$$

Для систем мобильной типовые значения спектральной эффективности лежат в диапазоне $~0,1 < \eta_s < 10~$ в зависимости от стандарта: 2G, 3G, 4G и 5G. Перевал за единицу произошел на стыке сетей третьего и четвертого поколений.

## Энергетическая эффективность систем связи.

Данная величина является оборотной стороной медали по отношению к рассмотренной выше спектральной эффективности:

$$\eta_e = V_b / P, ~бит/Дж.$$

Здесь $P$ – средняя потребляемая системой мощность. В контексте общей теории связи под средней мощностью понимают среднюю мощность, определяемую по сигнальному созвездию. Такая средняя мощность, в целом, пропорциональна порядку модуляции, то есть количеству символов созвездия:

$$P \sim M {d}^2_{\text {min}}$$

Здесь ${d}_{\text{min}}$ — минимальное расстояние между точками сигнального созвездия. Рост порядка модуляции улучшает спектральную эффективность, но ухудшает энергетическую.


## Ресурсная эффективность систем связи.

Взвешенное произведение спектральной и энергетической эффективностей дает ресурсную эффективность системы связи:

$$\eta_r = {\eta_s}^{w} {\eta_e}^{1 - w}$$

В зависимости от типа системы связи отдают приоритет либо спектральной эффективности, либо энергетической, регулируя параметр $0 \leq w \leq 1$. Современные мобильные системы связи (5G и 6G) могут регулировать этот параметр, например, в зависимости от времени суток: днем отдавая приоритет спектральной эффективности, а ночью — энергетической.

Иногда говорят об энерго-спектральной эффективности, когда находят «точку равновесия», при которой система не тратит лишнюю энергию ради незначительного прироста скорости.
